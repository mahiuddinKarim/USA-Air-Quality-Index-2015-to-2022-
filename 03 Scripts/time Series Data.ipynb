```python
# Total Distribution of Good AQI Values (0-50) in the United States
plot = sns.countplot(data = df, x = "AQI", color = "#00DB16")
plt.title("Overall Count of Good AQI Values (AQI 0-50) in the USA")
plot.set_xticklabels(plot.get_xticklabels(), fontsize = 9)
plt.xlim(0, 51)
plt.figure(figsize = (10, 12))
plt.show()
```

```python
#Total Distribution of Moderate AQI Values (50-100) in the United States
plot = sns.countplot(data = df, x = "AQI", color = "#F3FC00")
plt.title("Overall Count of Moderate AQI Values (AQI 50-100) in the USA")
plot.set_xticklabels(plot.get_xticklabels(), fontsize = 9)
plt.xlim(50, 101)
plt.figure(figsize = (10, 12))
plt.show()
```

```python
#Distribution of AQI Values (0-150) in Each State
for state in sorted(['Washington', 'Ohio', 'Georgia', 'Oregon', 'New York', 'New Mexico', 'Texas',
 'Michigan', 'Wisconsin', 'Oklahoma', 'Arkansas', 'Maine', 'California',
 'Maryland', 'Louisiana', 'Nebraska', 'Montana', 'North Dakota', 'Virginia',
 'Illinois', 'Indiana', 'Massachusetts', 'Colorado', 'Kentucky', 'Florida',
 'Utah', 'Iowa', 'West Virginia', 'South Carolina', 'North Carolina',
 'Tennessee', 'New Hampshire', 'Mississippi', 'Missouri', 'Alabama',
 'Minnesota', 'Arizona', 'Pennsylvania', 'Wyoming', 'Puerto Rico', 'Hawaii',
 'Idaho', 'Nevada', 'Rhode Island', 'New Jersey', 'District of Columbia',
 'South Dakota', 'Alaska', 'Connecticut', 'Vermont', 'Kansas', 'Delaware']):
    
    df0 = df[df["state_name"] == state]    
    
    plot = sns.countplot(data = df0, x = "AQI", color = "#00DB16")
    plt.title(f"Count of Good AQI Values (AQI 0-50) in {state}")
    plot.set_xticklabels(plot.get_xticklabels(), fontsize = 9)
    plt.xlim(0, 51)
    plt.show()
    
    df0 = df[df["state_name"] == state]
    
    plot = sns.countplot(data = df0, x = "AQI", color = "#F3FC00")
    plt.title(f"Count of Moderate AQI Values (AQI 50-100) in {state}")
    plot.set_xticklabels(plot.get_xticklabels(), fontsize = 9)
    plt.xlim(50, 100)
    plt.show()
    
    df0 = df[df["state_name"] == state]
    
    plot = sns.countplot(data = df0, x = "AQI", color = "#FFA200")
    plt.title(f"Unhealthy for Sensitive Groups AQI Values (AQI 100-150) in {state}")
    plot.set_xticklabels(plot.get_xticklabels(), rotation = 270, fontsize = 9)
    plt.xlim(100, 150)
    plt.show()
```
